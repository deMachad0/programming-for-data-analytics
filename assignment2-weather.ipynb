```python
# Creating a plot from a csv file
# Author: Andre Machado

import csv
import matplotlib.pyplot as plt
import numpy as np

FILENAME = "weatherreadings1.csv"
DATADIR = r"C:\Users\demac\OneDrive\Desktop\pands\programming-for-data-analytics\PFDA -assignments\\"

# Ensure the correct path with a separator between the directory and filename
filepath = DATADIR + FILENAME

total_dryBulbTemperature_Celsius = 0
count = 0
temperatures = []


try:
    with open(filepath, "rt") as fp:
        reader = csv.DictReader(fp)  # reads the first line automatically
        for row in reader:
            if "dryBulbTemperature_Celsius" in row:  # ensure column exists and is valid 
                total_dryBulbTemperature_Celsius += float(row["dryBulbTemperature_Celsius"])
                temperature = float(row["dryBulbTemperature_Celsius"])
                temperatures.append(temperature) # add to temperatures list
                total_dryBulbTemperature_Celsius == temperature
                count += 1
    if count > 0:
        average_dryBulbTemperature_Celsius = total_dryBulbTemperature_Celsius / count
        print(f"Average dryBulbTemperature_Celsius: {int(average_dryBulbTemperature_Celsius)}") # converting the result into an Int, because there is no temperature float
    else:
        print("No valid dryBulbTemperature_Celsius data available.")
    
    # use numpy to create indices array
    indices = np.arange(len(temperatures))

    fig ,ax = plt.subplots()
    ax.scatter(indices, temperatures, color='blue', label='Dry Bulb Temperature (Celsius)')
    ax.axhline(average_dryBulbTemperature_Celsius, color='red', linestyle='--', label=f'Average tempeture: {average_dryBulbTemperature_Celsius:.2f}C')
    ax.set_xlabel("Index")
    ax.set_ylabel("Dry Bulb Temperature (Â°C)")
    ax.set_title("Dry Bulb Temperature Readings")
    ax.legend()

    plt.show()

except FileNotFoundError: # in case of file not found
    print(f"Error: The file {filepath} was not found.")
```
